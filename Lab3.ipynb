```python
from pulp import *
a,b,c,d,e = 3,16,15,1,7
prob = LpProblem("Maximize problem", LpMaximize)
x1 = LpVariable("x1", lowBound=0)  
x2 = LpVariable("x2", lowBound=0)  
x3 = LpVariable("x3", lowBound=0)
x4 = LpVariable("x4", lowBound=0)
x5 = LpVariable("x5", lowBound=0)

prob += e * x1 + d * x2 - (b - a) * x3 - x4 + ( d - c) * x5 <= d + 2 * e +35
prob += (-a) * x1 -b * x2 - x3 + (d - c) * x4 + (a + b) * x5 <= 30 + e
prob += a * x1 + (c + e) * x2 + x3 + (e + b) * x4 + c * x5 <= b + 27

prob += (a - b + d) * x1 + (e - c ) * x2 + a * x3 - (b + d ) * x4 - (a + d - e) * x5

status = prob.solve()
prob.solve(PULP_CBC_CMD(msg=True))
print(f"Статус ЗЛП: {LpStatus[status]}")
print("Результат функції при максимізації:")
for i in prob.variables():
    print(i.name, '=', i.varValue)
print(f"Оптимальне значення цільової функції: z = {value(prob.objective)}")
```

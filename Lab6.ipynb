```python
import pulp
problem = pulp.LpProblem("ЦЗЛП", pulp.LpMaximize)
x1 = pulp.LpVariable('x1', lowBound=0, cat='Integer')
x2 = pulp.LpVariable('x2', lowBound=0, cat='Integer')
x3 = pulp.LpVariable('x3', lowBound=0, cat='Integer')
x4 = pulp.LpVariable('x4', lowBound=0, cat='Integer')
x5 = pulp.LpVariable('x5', lowBound=0, cat='Integer')
problem += -x1 + 8*x2 - x3 + x5, "Цільова функція"

problem += -2*x1 + 10*x2 + 3*x3 - 5*x4 + 2*x5 <= 3
problem += x1 + 3*x2 + 2*x3 + 2*x5 <= 5
problem += 4*x1 + 2*x2 + x3 - 3*x4 + 7*x5 <= 9
problem.solve()
print("Статус:", pulp.LpStatus[problem.status])
for variable in problem.variables():
    print(f"{variable.name} = {variable.varValue}")
print("Цільова функція:", pulp.value(problem.objective))
```

```python
import pulp as lp
import numpy as np
workers = ['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7']
tasks = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7']
matrix_c = {
    ('W1','T1'): 9,('W1','T2'): 2,('W1','T3'): 6,('W1','T4'): 13,('W1','T5'): 6,('W1','T6'): 2,('W1','T7'): 8,
    ('W2','T1'): 18,('W2','T2'): 4,('W2','T3'): 5,('W2','T4'): 16,('W2','T5'): 10,('W2','T6'): 12,('W2','T7'): 15,
    ('W3','T1'): 4,('W3','T2'): 7,('W3','T3'): 12,('W3','T4'): 13,('W3','T5'): 9,('W3','T6'): 11,('W3','T7'): 2,
    ('W4','T1'): 18,('W4','T2'): 11,('W4','T3'): 6,('W4','T4'): 10,('W4','T5'): 20,('W4','T6'): 20,('W4','T7'): 20,
    ('W5','T1'): 12,('W5','T2'): 7,('W5','T3'): 12,('W5','T4'): 20,('W5','T5'): 7,('W5','T6'): 17,('W5','T7'): 5,
    ('W6','T1'): 6,('W6','T2'): 12,('W6','T3'): 9,('W6','T4'): 18,('W6','T5'): 20,('W6','T6'): 20,('W6','T7'): 19,
    ('W7','T1'): 9,('W7','T2'): 2,('W7','T3'): 8,('W7','T4'): 2,('W7','T5'): 2,('W7','T6'): 13,('W7','T7'): 1
}

int_problem = lp.LpProblem('Assignment problem', lp.LpMinimize)
matrix_x = lp.LpVariable.dicts("X",
                               [(w,t) for w in workers for t in tasks],
                               cat='Binary')
int_problem += lp.lpSum([matrix_c[(w,t)] * matrix_x[(w,t)]
                         for w in workers for t in tasks])
for w in workers:
    int_problem += lp.lpSum([matrix_x[(w,t)] for t in tasks]) == 1
for t in tasks:
    int_problem += lp.lpSum([matrix_x[(w,t)] for w in workers]) == 1
print("Статус:", (int_problem.solve(lp.PULP_CBC_CMD(msg=0))))
print("Рішення:", lp.LpStatus[int_problem.status])
print(int_problem)
assignment_matrix = np.zeros((len(workers), len(tasks)), dtype=int)
for w in range(len(workers)):
    for t in range(len(tasks)):
        assignment_matrix[w, t] = int(matrix_x[(workers[w], tasks[t])].varValue)
print(assignment_matrix)
print("Значення цільової функції:", lp.value(int_problem.objective))
```

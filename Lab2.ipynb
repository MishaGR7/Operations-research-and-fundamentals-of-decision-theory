```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import numpy as np
from pulp import *
%config InlineBackend.figure_format = 'svg'
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = 5,5

prob = LpProblem("Minimize and maximize velue", LpMaximize)

probl = LpProblem("Minimize and maximize velue", LpMinimize)

x1 = LpVariable("x1", lowBound=0)  
x2 = LpVariable("x2", lowBound=0)  

prob += -3 * x1 + 4 * x2
prob += (5 * x1 + 7 * x2) >= 94
prob += (x1 + 13 * x2) <= 42 
prob += (-2 * x1 + 3 * x2) >= 3
status = prob.solve()
prob.solve(PULP_CBC_CMD(msg=True))
print(f"Статус ЗЛП: {LpStatus[status]}")
print("Результат функції при максимізації:")
for i in prob.variables():
    print(i.name, '=', i.varValue)
print(f"Оптимальне значення цільової функції: z = {value(prob.objective)}")
probl += -3 * x1 + 4 * x2
probl += (5 * x1 + 7 * x2) >= 94
probl += (x1 + 13 * x2) <= 42 
probl += (-2 * x1 + 3 * x2) >= 3
status1 = probl.solve()
probl.solve(PULP_CBC_CMD(msg=True))
print(f"Статус ЗЛП: {LpStatus[status1]}")
print("Результат функції при мінімізації:")
for i in probl.variables():
    print(i.name, '=', i.varValue)
print(f"Оптимальне значення цільової функції: z = {value(probl.objective)}")

x_1 = x1.value()
x_2 = x2.value()
x = np.arange(-20, 20)

x_fill = [-26,3,-6.6]
y_fill = [5.3,3,-3.7]
plt.fill(x_fill, y_fill, 'gray')
plt.plot(-3 * x, 4 * x, label='z = -3x1 + 4x2', color='red')
plt.plot(x, (94 - 7*x)/5, label='5x1 + 7x2 >= 94', color='blue')
plt.plot(x, (42 - x)/13, label='x1 + 13x2 <= 42,', color='green')
plt.plot(x, (-3 - 2*x)/(-3), label='-2x1 + 3x2 >= 3', color='purple')
plt.text(2.5,0.5, 'Багатокутник розв\'язків', size = '8')
plt.annotate(f'Оптимальний розв\'язок \n ({x_1}, {x_2})',size= '8',
             xy = (x_1,x_2))
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Графічний роз\'вязок')
plt.arrow(0,0,-3,4,
          length_includes_head=True,
          head_width=0.5,
          head_length=0.5,
          color='black')
plt.axis([-10,10,0,10])
plt.plot(4.4 * x - 3, 4 + x, 'g--')
plt.grid(True)
plt.legend()
plt.show()
```

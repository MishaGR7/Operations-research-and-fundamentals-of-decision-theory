```python
from pulp import *
prob = LpProblem("Max", LpMaximize)
  
x1 = LpVariable("В1_A1", lowBound=0)
x2 = LpVariable("В2_A1", lowBound=0)
x3 = LpVariable("В1_A2", lowBound=0)
x4 = LpVariable("В2_A2", lowBound=0)
prob += (2 * x1 + 3 * x3) + (2 * x2 + x4)

prob += 2 * x1 + 3 * x3 >= 950
prob += 2 * x2 + x4 >= 1200

prob += x1 + 5 * x2 <= 175
prob += 2 * x3 + x4 <= 160

prob += x1 >= 0
prob += x2 >= 0
prob += x3 >= 0
prob += x4 >= 0

prob.solve()
print("План графіку роботи:")
print(f"x1(B1_A1): {x1.varValue:.2f} годин роботи")
print(f"x2(B2_A1): {x2.varValue:.2f} годин роботи")
print(f"x3(B1_A2): {x3.varValue:.2f} годин роботи")
print(f"x4(B2_A2): {x4.varValue:.2f} годин роботи")
```

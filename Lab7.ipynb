```python
import pulp as pulp
A = [
    [5, 3, 4, 7],
    [5, 3, 9, 4],
    [4, 7, 6, 8]
]
prob = pulp.LpProblem("Maximize_v", pulp.LpMaximize)
x = [pulp.LpVariable(f'x{i}', lowBound=0) for i in range(3)]
v = pulp.LpVariable('v')
prob += v
for j in range(4):
    prob += sum(A[i][j] * x[i] for i in range(3)) >= v
prob += sum(x) == 1
prob.solve()
optimal_strategies_A = [pulp.value(var) for var in x]
game_value = pulp.value(v)
print(f"Оптимальні стратегії гравця A: {optimal_strategies_A}")
print(f"Ціна гри: {game_value}")
B = list(zip(*A))
prob_dual = pulp.LpProblem("Minimize_w", pulp.LpMinimize)
y = [pulp.LpVariable(f'y{j}', lowBound=0) for j in range(4)]
w = pulp.LpVariable('v')
prob_dual += w
for i in range(3):
    prob_dual += sum(B[j][i] * y[j] for j in range(4)) <= w
prob_dual += sum(y) == 1
prob_dual.solve()
optimal_strategies_B = [pulp.value(var) for var in y]
print(f"Оптимальні стратегії гравця B: {optimal_strategies_B}")
print(f"Ціна гри: {pulp.value(w)}")
```

```python
from pulp import *
a,b,c,d,e = 3,16,15,1,7
prob = LpProblem("Maximize problem", LpMaximize)
x1 = LpVariable("x1", lowBound=0)  
x2 = LpVariable("x2", lowBound=0)  
x3 = LpVariable("x3", lowBound=0)
x4 = LpVariable("x4", lowBound=0)
x5 = LpVariable("x5", lowBound=0)

prob += 9 * x1 - 16 * x2 - x3 <= 111
prob += 16 * x1 + 5 * x2 -4 * x3 <= 56
prob += (-3) * x2 - 12 * x3 <= 17

prob += 12 * x1 - 16 * x2 + 2 * x3

status = prob.solve()
prob.solve(PULP_CBC_CMD(msg=True))
print(f"Статус ЗЛП: {LpStatus[status]}")
print("Результат функції при максимізації:")
for i in prob.variables():
    print(i.name, '=', i.varValue)
print(f"Оптимальне значення цільової функції: z = {value(prob.objective)}")
```

```python
import pulp
prob = pulp.LpProblem("ZLP", pulp.LpMaximize)

x1 = pulp.LpVariable('x1', lowBound=0)
x2 = pulp.LpVariable('x2', lowBound=0)

prob += 3 * x1 + 2 * x2

prob += 11 * x1 + 6 * x2 <= 42
prob += 5 * x1 + 29 * x2 <= 63
prob.solve()
print("Статус:", pulp.LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Оптимальне значення цільової функції = ", pulp.value(prob.objective))
```

```python
import pulp
dual_prob = pulp.LpProblem("Dual_ZLP", pulp.LpMinimize)
y1 = pulp.LpVariable('y1', lowBound=0)
y2 = pulp.LpVariable('y2', lowBound=0)
y3 = pulp.LpVariable('y3', lowBound=0)
y4 = pulp.LpVariable('y4', lowBound=0)

dual_prob += 42 * y1 + 63 * y2 + y3 + 2 * y4
dual_prob += 11 * y1 + 5 * y2 >= 3
dual_prob += 6 * y1 + 29 * y2 + y3 + y4 >= 2
dual_prob.solve()
print("Статус:", pulp.LpStatus[dual_prob.status])
for v in dual_prob.variables():
    print(v.name, "=", v.varValue)
print("Оптимальне значення цільової функції = ", pulp.value(dual_prob.objective))
```
